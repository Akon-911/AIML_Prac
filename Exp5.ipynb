```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as cnv
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import sklearn.metrics as metrics
```

```python
df = pd.read_csv("colon.csv").drop("Unnamed: 0", axis=1).dropna()
df
```

```python
df.drop(["id"], axis=1, inplace=True)
```

```python
df
```

```python
df.info()
```

```python
x = df.drop("etype", axis=1)
y = df["etype"]
```

```python
y
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=7)
```

```python
x_train
```

```python
x_test
```

```python
LModel = LogisticRegression(max_iter=1000)
```

```python
LModel.fit(x_train,y_train);
```

```python
y_predict = LModel.predict(x_test)
```

```python
model_score = LModel.score(x_test,y_test)
```

## Confusion Metrics

```python
metrics.confusion_matrix(y_test,y_predict)
```

## Classification Report

```python
print(metrics.classification_report(y_test, y_predict))
```
