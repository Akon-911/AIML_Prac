```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

## Importing Sample Data sets from sklean's premade datasets

```python
from sklearn.datasets import load_breast_cancer
dataToUse = load_breast_cancer()
```

```python
dataToUse.keys()
```

## target_names => Output Classes
## feature_names => Inputs

```python
dataToUse["target_names"]
```

```python
dataToUse["feature_names"]
```

```python
DF1 = pd.DataFrame(dataToUse["data"], columns=dataToUse["feature_names"])
```

```python
DF1.info()
```

### Features -> 30

Now here, we will scale the data scaling

```python
scaler = StandardScaler()

scaler.fit(DF1)
```

```python
scaled_data = scaler.transform(DF1)
```

```python
scaled_data.shape
```

### Now it's similar to numpy Array form

```python
principal = PCA(n_components=3)
```

```python
principal.fit(scaled_data)
```

```python
x = principal.transform(scaled_data)
```

```python
x
```

```python
x.shape
```

### No. of features reduced from 30 to 3

```python
principal.components_
```

```python
plt.scatter(x[:,0],x[:,1],c=dataToUse['target'],cmap='plasma')
plt.xlabel("PC1")
plt.ylabel("PC2")
```

```python
plt.scatter(x[:,0],x[:,2], c=dataToUse["target"],cmap="plasma")
plt.xlabel("PC1")
plt.ylabel("PC3")
```
