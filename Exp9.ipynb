```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.naive_bayes import GaussianNB
```

```python
data_RAW = load_iris()
```

```python
data = pd.DataFrame(data_RAW.data, columns=data_RAW.feature_names)
```

```python
data
```

```python
data["color"] = data_RAW.target
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_Train, X_Test, Y_Train, Y_Test = train_test_split(data.drop("color",axis=1), data["color"], test_size=0.3)
```

```python
X_Train,X_Test
```

```python
Y_Train, Y_Test
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(X_Train, Y_Train)
```

```python
Y_Pred = gnb.predict(X_Test)
```
